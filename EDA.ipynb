```python
import torch
import matplotlib.pyplot as plt

# train_X.pt
train_X = torch.load('data/train_X.pt')

# 10
fig, axes = plt.subplots(2, 1, figsize=(10, 10))  # 101


for i in range(5):
    axes[0].plot(train_X[0][i].numpy())  # i
    axes[0].set_title(f'Train Dataset X[{i}]')
    axes[0].set_xlabel('Time')
    axes[0].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
# 301CH
fig, axes = plt.subplots(30, 1, figsize=(10, 60))  # 301

for i in range(30):
    axes[i].plot(train_X[i][0].numpy())  # i1CH
    axes[i].set_title(f'Train Dataset X[{i}] - Channel 1')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
