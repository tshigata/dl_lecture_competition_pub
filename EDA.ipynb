```python
import torch
import matplotlib.pyplot as plt

# train_X.ptファイルの読み込み
train_X = torch.load('data/train_X.pt')

# 先頭10個の波形データを一度に表示
fig, axes = plt.subplots(2, 1, figsize=(10, 10))  # 10行1列のサブプロット


for i in range(5):
    axes[0].plot(train_X[0][i].numpy())  # i番目の波形データをプロット
    axes[0].set_title(f'Train Dataset X[{i}]')
    axes[0].set_xlabel('Time')
    axes[0].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

```python
# 先頭30サンプルの1CHのデータを一度に表示
fig, axes = plt.subplots(30, 1, figsize=(10, 60))  # 30行1列のサブプロット

for i in range(30):
    axes[i].plot(train_X[i][0].numpy())  # i番目のサンプルの1CHデータをプロット
    axes[i].set_title(f'Train Dataset X[{i}] - Channel 1')
    axes[i].set_xlabel('Time')
    axes[i].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
```
